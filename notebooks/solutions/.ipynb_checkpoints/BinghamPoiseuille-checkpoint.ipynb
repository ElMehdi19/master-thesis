```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
dpdz = -4
R = 2
yield_stress = 2
mu = 4 * pow(10, -3)
# r_c = 1
```

```python
@np.vectorize
def newtonian(r):
    return dpdz * (r**2 - R**2) / (4 * mu)

@np.vectorize
def bingham(r, ys=yield_stress, dpdz=dpdz):
    r_c = -2 * ys / dpdz
    if abs(r) >= 0 and abs(r) <= r_c:
        return 2 * newtonian(r_c) - 2 * ys * (r_c - R) / mu
    if r < 0:
        r *= - 1
    return 2 * newtonian(r) - 2 * ys * (r - R) / mu
```

```python
r = np.linspace(-R, R, 150)
b_y05 = bingham(r, ys=0.5)
b_y1 = bingham(r, ys=1)
b_y2 = bingham(r, ys=2)
b_y3 = bingham(r, ys=3)
```

```python
with plt.style.context('fivethirtyeight'):
#     plt.plot(r, newtonian(r), label='Newtonian', linewidth=1)
#     plt.plot(b_y05, r, linewidth=1, label=r'$\sigma_{c}=0.5$')
    plt.plot(b_y1, r, linewidth=1, label=r'$\sigma_{c}=1$')
    plt.plot(b_y2, r, linewidth=1, label=r'$\sigma_{c}=2$')
    plt.plot(b_y3, r, linewidth=1, label=r'$\sigma_{c}=3$')
#     plt.plot(b_y4, r, linewidth=1, label=r'$\sigma_{c}=4$')
    plt.ylabel('$r$')
    plt.xlabel('$w(r)$')
    plt.legend(framealpha=0.1, edgecolor='black')
    sp = r'\hspace{0.25}'
    frac = r'\frac{dP}{dz}'
    sig = r'\sigma_{c}'
    ttl = fr'$Repr√©sentation{sp}graphique{sp}de{sp}w(r){sp}pour{sp}{frac}=-4,{sp}\mu=0.4$'
    plt.title(ttl, fontsize=14)
    plt.tight_layout()
    fig = plt.gcf()
    fig.set_size_inches(12.5, 5.5)
    plt.savefig('figs/w-bingham-poiseuille.png', transparent=True, dpi=100)
```

```python
r = np.linspace(-R, R, 150)
b_dp_4 = bingham(r)
b_dp_2 = bingham(r, dpdz=-3)
b_dp_8 = bingham(r, dpdz=-8)
# b_y2 = bingham(r, ys=2)
# b_y3 = bingham(r, ys=3)
```

```python
with plt.style.context('fivethirtyeight'):
    plt.plot(b_dp_2, r, linewidth=1, label=r'$\frac{dP}{dz}=-2$', linestyle='dashed')
    plt.plot(b_dp_4, r, linewidth=1, label=r'$\frac{dP}{dz}=-4$')
    plt.plot(b_dp_8, r, linewidth=1, label=r'$\frac{dP}{dz}=-8$', linestyle='None', marker='.', color='green')

    
    plt.ylabel('$r$')
    plt.xlabel('$w(r)$')
    plt.legend(framealpha=0.1, edgecolor='black')
    sp = r'\hspace{0.25}'
    frac = r'\frac{dP}{dz}'
    sig = r'\sigma_{c}'
    ttl = fr'$Repr√©sentation{sp}graphique{sp}de{sp}w(r){sp}pour{sp}{sig}=2,{sp}\mu=0.4$'
    plt.title(ttl, fontsize=14)
    plt.tight_layout()
    fig = plt.gcf()
    fig.set_size_inches(12.5, 5.5)
    plt.savefig('figs/w-bingham-poiseuille-dpdz.png', transparent=True, dpi=100)
```

```python
def output(dpdz):
    part_1 = -np.math.pi * dpdz * np.math.pow(R, 4) / (4 * mu)
    part_2 = 4 * np.math.pi * yield_stress * pow(R, 3) / (6 * mu)
    return part_1 + part_2

output = np.vectorize(output)
```

```python
dp = np.linspace(-10, 0, 50)
with plt.style.context('fivethirtyeight'):
    plt.plot(dp, output(dp), linewidth=1)
    plt.xlabel(r'$\frac{dp}{dz}$')
    plt.ylabel(r'$Q(\frac{dp}{dz})$')
#     plt.legend(framealpha=0.1, edgecolor='black', bbox_to_anchor=(1, 0.5))
    sp = r'\hspace{0.25}'
    frac = r'\frac{\partial P}{\partial z}'
    sig = r'\sigma_{c}'
    ttl = fr'$Repr√©sentation{sp}graphique{sp}de{sp}Q{sp}pour{sp}{sig}=0.1,{sp}{frac}=-4,{sp}\mu=0.4$'
    plt.title(ttl, fontsize=14)
    plt.tight_layout()
    fig = plt.gcf()
    fig.set_size_inches(12.5, 5.5)
    plt.savefig('figs/q-bingham-poiseuil.png', transparent=True, dpi=100)
```

```python
@np.vectorize
def critical_r(ys=yield_stress, dpdz=-dpdz):
    return 2 * ys / dpdz
```

```python
dp = np.linspace(-10, 0, 50)
with plt.style.context('fivethirtyeight'):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
    
    sp = r'\hspace{0.25}'
    frac = r'\frac{dP}{dz}'
    sig = r'\sigma_{c}'
    rc = r'r_{c}'
    
    ax1.plot(np.linspace(1, 10, 100), critical_r(ys=np.linspace(1, 10, 100)))
    ax1.set_xlabel(r'$\sigma_{c}$')
    ax1.set_ylabel(r'$r_{c}$')
    ax1.set_title(fr'$Repr√©sentation{sp}graphique{sp}de{sp}{rc}{sp}pour{sp}{frac}=2,{sp}\mu=0.4$', fontsize=14)

    ax2.plot(np.linspace(-10, -1, 100), critical_r(dpdz=np.linspace(-10, -1, 100)))
    ax2.set_xlabel(r'$\frac{dp}{dz}$')
    ax2.set_ylabel(r'$r_{c}$')
    ax2.set_title(fr'$Repr√©sentation{sp}graphique{sp}de{sp}{rc}{sp}pour{sp}{sig}=2,{sp}\mu=0.4$', fontsize=14)

    plt.tight_layout()
    fig = plt.gcf()
    fig.set_size_inches(12.5, 5.5)
    plt.savefig('figs/bingham-poiseuille-rc.png', transparent=True, dpi=100)
```
