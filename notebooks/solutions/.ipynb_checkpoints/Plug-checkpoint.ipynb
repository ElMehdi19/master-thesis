```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
zs = np.linspace(0, 4, 200)
```

```python
@np.vectorize
def r2(z, eps=0.1):
    return 1 + eps * np.math.cos(np.math.pi * z)
```

```python
with plt.style.context('fivethirtyeight'):
    fig = plt.gcf()
    fig.set_size_inches(12.5, 5.5)
    fig.suptitle('Representation des differentes phases de l\'Ã©coulement')
    
    plt.plot(zs, np.zeros(zs.shape), color='black', linewidth=1)
    plt.plot(zs, r2(zs), color='black', linewidth=1)
    plt.plot(zs, np.ones(zs.shape) * 0.3, color='black', linewidth=1, linestyle='dashed')
    plt.plot(zs, np.ones(zs.shape) * 0.6, color='black', linewidth=1, linestyle='dashed')
    
    # plug region lower-upper bounds
    plt.text(4.1, 0.3, r'$r=\lambda_{1}$')
    plt.text(4.1, 0.6, r'$r=\lambda_{2}$')
    
    # plug region stress tensor
    plt.text(1.8, 0.4, r'$\sigma_{i,j}<\sigma_{c}$', fontsize=20)
    
    plt.text(1.8, 0.1, r'$\sigma_{i,j}>\sigma_{c}$', fontsize=20)
    plt.text(1.8, 0.7, r'$\sigma_{i,j}>\sigma_{c}$', fontsize=20)
    plt.grid(False)
    
    plt.savefig('figs/plug-region.png', transparent=True)
```
