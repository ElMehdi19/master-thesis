```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import iv, kv
from PIL import Image
```

```python
I = lambda x: iv(0, x)
K = lambda x: kv(0, x)
```

```python
Image.open('newtonian.png')
```

```python
r1, r2 = 0.5, 1.5
M = 10
m = 1.5
A2 = 2 * pow(M, 2) / (1 + pow(m, 2))
```

```python
def c(dpdz):
    A = np.sqrt(A2)
    part_1 = 2 * dpdz / A2 - 1
    part_2 = (K(A * r2) - K(A * r1)) / (I(A * r1) - I(A * r2))
    part_2 = part_2 * I(A * r1) + K(A * r1)
    return part_1 * np.math.pow(part_2, -1)

c = np.vectorize(c)

def b(dpdz):
    A = np.sqrt(A2)
    return c(dpdz) * (K(A * r2) - K(A * r1)) / (I(A * r1) - I(A * r2))

def w(r, dpdz):
    A = np.sqrt(A2)
    return b(dpdz) * I(A * r) + c(dpdz) * K(A * r) - 2 * dpdz / A2
```

```python
r = np.linspace(r1, r2, 100)
plt.plot(r, w(r, 60), label=r'$\frac{dp}{dz}=60$')
plt.plot(r, w(r, 100), label=r'$\frac{dp}{dz}=100$')
plt.legend()
```

```python
Image.open('discretized.png')
```

```python
def rhs(t, w, N, M, m, dpdz):
    dwdt = np.zeros(N)
    w[0], w[-1] = -1, -1
    dr = 1 / N
    
    try:
#         print(t)
        dpdz = dpdz(t)
    except TypeError:
        dpdz = dpdz
        
    for i in range(1, N - 1):
        r = i * dr + 0.5
        wrr = (w[i + 1] - 2 * w[i] + w[i - 1]) / (dr ** 2)
        wr = (w[i + 1] - w[i - 1]) / (dr * 2)
        A = 0.5 * (wrr + wr / r)
        B = pow(M, 2) * (w[i] + 1) / (1 + pow(m, 2))
        dwdt[i] = A - dpdz - B
    
    return dwdt
```

```python
import pandas as pd
from scipy.integrate import solve_ivp, odeint
```

```python
N = 100
M = 0
m = 1.5
dpdz = 10

w0 = np.empty(N)
w0[:] = -1
# s1 = solve_ivp(rhs, (0, 1), w0, args=(N, M, m, dpdz))
t = np.linspace(0, 1, 1000)
s1 = odeint(rhs, w0, t, args=(N, M, m, dpdz), tfirst=True)
df1 = pd.DataFrame(s1)
```

```python
zipped = zip(np.arange(0, 100, 20), t[:5])
r = np.array([i / N + 0.5 for i in range(0, N)])

with plt.style.context('fast'):
    for (i, x) in zipped:
        plt.plot(r, df1.loc[i], linewidth=2, label=f'$t={x * pow(10, 3):.3f} (ms)$')
    plt.legend()
    plt.xlim(left=0.5, right=1.5)
    plt.grid()
    plt.title(r'Axial velocity for different values of $t$ and $\frac{dp}{dz}=10$')
    plt.xlabel('$r$', fontsize=14)
    plt.ylabel('$w$', fontsize=14)
```
