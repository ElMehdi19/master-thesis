```python
def rhs(w, t, N, M, m, Bn, dpdz):
    dwdt = np.zeros(N)
    w[0], w[-1] = 0, -1
    dr = 1 / N
    
    for i in range(1, N-1):
        wr = (w[i+1] - w[i-1]) / (2 * dr)
        wrr = (w[i+1] - 2 * w[i] + w[i-1]) / (dr ** 2)
        r = i * dr + 0.5
        A = 0.5 * (wrr + wr / r)
        B = (M**2) * (w[i] + 1) / (1 + m**2)
        C = Bn / r
        dwdt[i] = A - B + C - dpdz
    
    return dwdt
```

```python
N = 10
M = 100
m = 1.5
Bn = 0.02
dpdz = 0.3
t = [0, 1, 2, 3, 4]
w0 = np.zeros(N)
w0[-1] = -1
sol = odeint(rhs, w0, t, args=(N, M, m, Bn, dpdz))
```

```python
from scipy.special import kv, iv
I = lambda x: iv(0, x)
K = lambda x: kn(0, x)
```

```python
r1, r2 = 0.5, 1.5
```

```python
def c(dpdz, A2):
    A = np.sqrt(A2)
    term_1 = (2 * dpdz / A2) * (1 - I(A * r1) / I(A * r2)) + 1
    term_2 = K(A * r1) - K(A * r2) * I(A * r1) / I(A * r2)
    return term_1 * np.math.pow(term_2, -1)

def b(dpdz, A2):
    A = np.sqrt(A2)
    term_1 = 2 * dpdz / (A2 * I(A * r2))
    term_2 = c(dpdz, A2) * K(A * r2) / I(A * r2)
    return term_1 - term_2

def w(r, dpdz, A2):
    A = np.sqrt(A2)
    return b(dpdz, A2) * I(A * r) + c(dpdz, A2) * K(A * r) - (2 * dpdz / A2) - 1
w = np.vectorize(w)
```
