```python
import numpy as np
from scipy.integrate import RK45, solve_ivp
from matplotlib import pyplot as plt
```

```python
def rhs(w, t, N, M, m, Bn, dpdz):
    dwdt = np.zeros(N)
    w[0], w[-1] = 0, -1
    dr = 1 / N
    
    for i in range(1, N-1):
        wr = (w[i+1] - w[i-1]) / (2 * dr)
        wrr = (w[i+1] - 2 * w[i] + w[i-1]) / (dr ** 2)
        r = i * dr
        A = 0.5 * (wrr + wr / r)
        B = (M**2) * (w[i] + 1) / (1 + m**2)
        C = Bn / r
        dwdt[i] = A - B + C - dpdz
    
    return dwdt
```

```python
N = 40
```

```python
import pandas as pd
```

```python
def solver_1(t, w):
    M = 100
    m = 1.5
    Bn = 100
    dpdz = 0.3
    return rhs(w, t, N, M, m, Bn, dpdz)

w0 = np.zeros(N)
w0[-1] = 0

s1 = solve_ivp(solver_1, (0, 5), w0, max_step=1)
df = pd.DataFrame(s1.y)
```

```python
plt.style.use('fivethirtyeight')
# for i in range(20):
#     plt.plot(np.arange(N), df.iloc[:, i], label=f'$t={i}$', linewidth=2)
plt.plot(np.arange(N), df.iloc[:, 10])
```

```python
def solver_2(t, w):
    M = 100
    m = 1.5
    Bn = 100
    dpdz = 100
    return rhs(w, t, N, M, m, Bn, dpdz)

s2 = solve_ivp(solver_2, (0, 5), w0, max_step=1)
df2 = pd.DataFrame(s2.y)
```

```python
plt.style.use('fivethirtyeight')
for i in range(8):
    plt.plot(np.arange(N), df2.iloc[:, i], label=f'$t={i}$', linewidth=2)
```
