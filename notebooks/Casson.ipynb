```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
sig_c = 10
visco = 10

def casson(x):
    return pow(np.sqrt(sig_c) + np.sqrt(visco * x), 2)
CASSON = np.vectorize(casson)
```

```python
def bi_visco(x):
    pi = x * x
#     pic = 2
#     pi = pi if pi > pic else pic
    return 2 * (visco + sig_c / np.sqrt(2 * pi)) * x
BI = np.vectorize(bi_visco)
```

```python
x = np.linspace(0.1, 2, 2000)
y1 = np.array([CASSON(n) for n in x])
y2 = np.array([BI(n) for n in x])
```

```python
with plt.style.context('fast'):
    # CASSON
    plt.plot(x, y1, linewidth=1, linestyle='solid', color='green', label='Casson')
    plt.plot(x, y2, linewidth=1, linestyle='solid', color='red', label='Bi-Viscous')
    plt.xlabel('dij')
    plt.ylabel('Sigma')
    plt.xlim(left=0.1)
    plt.ylim(bottom=0)
    plt.tight_layout()
    plt.legend()
    plt.show()
```
