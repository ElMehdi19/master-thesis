```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
viscosity = 0.012

def sisko(shear_rates, k=1, n=1):
    return viscosity * shear_rates + k * np.math.pow(shear_rates, n)

sisko = np.vectorize(sisko)
shear_rates = np.linspace(0, 2, 300)
law = '$\sigma_{ij}=\mu_{0}d_{ij}+kd^{n}_{ij}$'
```

```python
with plt.style.context('fivethirtyeight'):
    for k in np.arange(0.2, 1, 0.2):
        plt.plot(shear_rates, sisko(shear_rates, k=k), label=f'$k={k:.2f}$', linewidth=2)
    plt.xlabel('$d_{ij}$')
    plt.ylabel('$\sigma_{ij}$')
    plt.legend()
    plt.tight_layout()
    plt.title(f'Sisko rheogram for $n=1$ and various values of $k$\n{law}', fontsize=14, y=-0.4)
    plt.show()
```

```python
with plt.style.context('fivethirtyeight'):
    for n in np.arange(0.2, 1, 0.2):
        plt.plot(shear_rates, sisko(shear_rates, n=n), label=f'$n={n:.2f}$', linewidth=2)
    plt.xlabel('$d_{ij}$')
    plt.ylabel('$\sigma_{ij}$')
    plt.legend()
    plt.tight_layout()
    plt.title(f'Sisko rheogram for $k=1$ and various values of $n$\n{law}', fontsize=14, y=-0.4)
    plt.show()
```
