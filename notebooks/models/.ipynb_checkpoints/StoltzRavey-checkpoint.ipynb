```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
viscosity = 0.012
def sloltz_ravey(shear_rates, a=1, b=1):
    shear_sqrt = np.sqrt(shear_rates)
    numerator = np.math.pow(1 + a * shear_sqrt, 2)
    denominator = np.math.pow(1 + b * shear_sqrt, 2)
    return viscosity * shear_rates * numerator / denominator

sloltz_ravey = np.vectorize(sloltz_ravey)
shear_rates = np.linspace(0, 2, 300)
```

```python
with plt.style.context('fivethirtyeight'):
    for a in np.arange(0, 1, 0.2):
        plt.plot(shear_rates, sloltz_ravey(shear_rates, a=a), label=f'$a={a:.2f}$', linewidth=2)
    plt.xlabel('$d_{ij}$')
    plt.ylabel('$\sigma_{ij}$')
    plt.tight_layout()
    plt.title(f'Stoltz Ravey rheogram for $b=1$ and various values of $a$', fontsize=14, y=-0.3)
    plt.legend()
    plt.show()
```

```python
with plt.style.context('fivethirtyeight'):
    for b in np.arange(0, 1, 0.2):
        plt.plot(shear_rates, sloltz_ravey(shear_rates, b=b), label=f'$b={b:.2f}$', linewidth=2)
    plt.xlabel('$d_{ij}$')
    plt.ylabel('$\sigma_{ij}$')
    plt.tight_layout()
    plt.title(f'Stoltz Ravey rheogram for $a=1$ and various values of $b$', fontsize=14, y=-0.3)
    plt.legend()
    plt.show()
```
