```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
sig_c = 10
visco = 10

def casson(x):
    return pow(np.sqrt(sig_c) + np.sqrt(visco * x), 2)
CASSON = np.vectorize(casson)
```

```python
def bi_visco(x):
    pi = x * x
#     pic = 2
#     pi = pi if pi > pic else pic
    return 2 * (visco + sig_c / np.sqrt(2 * pi)) * x
BI = np.vectorize(bi_visco)
```

```python
x = np.linspace(0.1, 2, 2000)
y1 = CASSON(x)
#y2 = np.array([BI(n) for n in x])
```

```python
with plt.style.context('seaborn'):
    # CASSON
    plt.plot(x, y1, linewidth=2, linestyle='solid', color='blue', label='Casson')
    #plt.plot(x, y2, linewidth=1, linestyle='solid', color='red', label='Bi-Viscous')
    plt.xlabel('dij', fontsize=18, x=1)
    plt.ylabel(u'$\sigma$', rotation=0, labelpad=0, y=0.9, horizontalalignment='right', fontsize=18)
    plt.xlim(left=0.1)
    plt.ylim(bottom=0)
    plt.tick_params(axis='both', which='both', left=False, bottom=False, labelbottom=False, labelleft=False)
    #plt.tight_layout()
    plt.title('Figure 1: Rh√©ogramme d\'un fluide de Casson', y=-0.12, fontsize=15, color='blue')
    plt.legend()
    plt.savefig(os.path.abspath('../../playground/figs/fig1.png'))
    plt.show()
```
