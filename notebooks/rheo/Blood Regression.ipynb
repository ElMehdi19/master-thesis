```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_2.csv')
df = df.head(13)
```

```python
from matplotlib import pyplot as plt
with plt.style.context('fivethirtyeight'):
    plt.scatter(df.ShearRate, df.Stress)
    plt.xlabel('$d_{ij} (s^{-1})$')
    plt.ylabel('$\sigma_{ij} (Pa)$')
    plt.tight_layout()
    plt.savefig('fig_scatter_data.png', transparent=True)
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression(fit_intercept=False)
reg.fit(df[['ShearRate']], df.Stress)
```

```python
import numpy as np
a, b = reg.coef_[0] / 2, reg.intercept_
def newtonian(shear_rates):
    return 2 * a * shear_rates
newtonian = np.vectorize(newtonian)
shear_rates = np.linspace(0, 100, 200)
```

```python
with plt.style.context('fivethirtyeight'):
    plt.plot(shear_rates, newtonian(shear_rates), linewidth=1, label='Newtonien', color='green')
    plt.scatter(df.ShearRate, df.Stress, label='Samples')
    plt.legend()
```

```python
a, b
```

```python
reg_bingham = LinearRegression()
reg_bingham.fit(df[['ShearRate']], df.Stress)
a_bingham, b_bingham = reg_bingham.coef_[0], reg_bingham.intercept_

def bingham(shear_rates):
    return a_bingham * shear_rates + b_bingham

bingham = np.vectorize(bingham)
```

```python
with plt.style.context('fivethirtyeight'):
    plt.plot(shear_rates, bingham(shear_rates), linewidth=1, label='Bingham', color='red')
    plt.scatter(df.ShearRate, df.Stress, label='Samples')
    plt.legend()
```

```python
a_bingham, b_bingham
```

```python
casson_reg = LinearRegression()
casson_reg.fit(df[['ShearRate']] ** 0.5, df.Stress ** 0.5)
a_casson, b_casson = casson_reg.coef_[0], casson_reg.intercept_

def casson(shear_rates):
    return np.math.pow(a_casson * np.math.pow(shear_rates, 0.5) + b_casson, 2)

casson = np.vectorize(casson)
```

```python
with plt.style.context('fivethirtyeight'):
    plt.plot(shear_rates, casson(shear_rates), linewidth=1, label='Casson', color='orange')
    plt.scatter(df.ShearRate, df.Stress, label='Samples')
    plt.legend()
```
